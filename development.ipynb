```python
#libraries
import pandas as pd
from datetime import date
from datetime import datetime
from datetime import timedelta
```

```python
#function to create inventory csv
def create_inventory_csv():
    inventory = pd.DataFrame(columns=['PRODUCT CODE' , 'BRAND', 'MODEL' , 'QUANTITY' , 'DECRIPTION', 'PLACE OF STORAGE'])
    inventory.to_csv('INVENTORY.csv', index=False) 
#function to open inventory csv
def open_inventory_csv():
    csv=pd.read_csv('INVENTORY.csv')
    return csv
#check and open inventory csv
def open_inventory_of_csv():
    try :
        csv=open_inventory_csv()
    except FileNotFoundError : 
        create_inventory_csv()
        csv=open_inventory_csv()
    return csv
#saving inventory csv
def save_inventory_of_csv(inventory_csv):
    inventory_csv.to_csv('INVENTORY.csv',index=False)
```

```python
#function to create sales csv
def create_sales_csv():
    sales = pd.DataFrame(columns=['PRODUCT CODE','COSTUMER NAME','BRAND', 'MODEL' , 'QUANTITY' ,'PLACE OF STORAGE','DATE','TIME','SELLING PRICE','BILLING CODE'])
    sales.to_csv('SALES.csv', index=False) 
#function to open sales csv
def open_sales_csv():
    csv=pd.read_csv('SALES.csv')
    return csv
#check and open sales csv
def open_sales_of_csv():
    try :
        csv=open_sales_csv()
    except FileNotFoundError : 
        create_sales_csv()
        csv=open_sales_csv()
    return csv
#saving sales csv
def save_sales_of_csv(sales_csv):
    sales_csv.to_csv('SALES.csv',index=False)
```

```python
#main menu 
def main_menu():
    print('1- INVENTORY MANAGER')
    print('2- SALES MANAGER')
    print('3- INVENTORY VIEW')
    print('4- SALES VIEW')
    print('5- INVENTORY ALERTS')
    print('6- EXIT')
    choice = 7
    while(choice>6):
        print('ENTER YOUR CHOICE : ')
        choice = int(input())
        if choice>6 : 
            print('ENTER A VALID CHOICE')
    return choice
    
```

```python
#inventory quantity changer
def quantity_changer(operation,quantity,product_code):
    inventory_csv=open_inventory_of_csv()
    quantity_list=list(inventory_csv['QUANTITY'])
    product_code_list=list(inventory_csv['PRODUCT CODE'])
    if operation=='add':
        quantity_list[product_code]=quantity+quantity_list[product_code]
    elif operation=='subtract':
        quantity_list[product_code]=quantity_list[product_code]-quantity
    inventory_csv['QUANTITY']=quantity_list
    save_inventory_of_csv(inventory_csv)
```

#inventory management inner functions -


```python
def inventory_manager_menu():
    print('1-SEARCH ITEM')
    print('2-ADD ITEM')
    print('3-MAIN MENU')
    print('ENTER YOUR CHOICE : ')
    choice = int(input())
    return choice
def inventory_search():
    inventory_csv=open_inventory_of_csv()
    print('ENTER THE MODEL NUMBER : ')
    model = input()
    model_list = inventory_csv[inventory_csv['MODEL']==model]
    print(model_list)
def existing_check(brand,model,place):
    inventory_csv=open_inventory_of_csv()
    search=inventory_csv[inventory_csv['BRAND']==brand]
    search=search[search['MODEL']==model]
    search=search[search['PLACE OF STORAGE']==place]
    return search
def inventory_add_item():
    inventory_csv=open_inventory_of_csv()
    print('ENTER THE BRAND : ')
    brand = input()
    print('ENTER THE MODEL NUMBER : ')
    model = input()
    print('ENTER THE QUANTITY : ')
    quantity = int(input())
    print('ENTER THE DISCRIPTION : ')
    description = input()
    print('ENTER THE PLACE OF STORAGE : ')
    place = input()
    product_code=len(inventory_csv)
    #limit=
    entry = [product_code , brand , model , quantity , description , place]
    check=existing_check(brand,model,place)
    if len(check)==0:
        inventory_csv.loc[len(inventory_csv)+1] = entry
        inventory_csv.reset_index(drop = True)
        inventory_csv.reset_index(drop = True)
        save_inventory_of_csv(inventory_csv)
    else :
        product_code=int(check['PRODUCT CODE'])
        quantity_changer('add',quantity,product_code)
```

#sales manager inner functions

```python
def get_date():
    today = date.today()
    d4 = today.strftime("%b-%d-%Y")
    return d4
def get_time():
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    return current_time
def sales_inventory_check(brand,model,place,quantity):
    inventory_csv=open_inventory_of_csv()
    search=inventory_csv[inventory_csv['BRAND']==brand]
    search=search[search['MODEL']==model]
    search=search[search['PLACE OF STORAGE']==place]
    search=search[search['QUANTITY']>=quantity]
    return search
def billing_code_generator(name,quantity,product_code):
    t = date.today()
    d4 = t.strftime("%b%d%Y")
    code = str(quantity) + name[0:2]  + d4 + str(product_code) 
    return code.upper()
def sales_manager_menu():
    print('1-ADD SALE')
    print('2-RETURN')    
    print('3-MAIN MENU') 
    print('ENTER YOUR CHOICE : ')
    choice = int(input())
    return choice
def add_sales():
    inventory_csv=open_inventory_of_csv()
    sales_csv=open_sales_of_csv()
    print('ENTER THE COSTUMER NAME : ')
    name=input()
    print('ENTER THE BRAND : ')
    brand = input()
    print('ENTER THE MODEL NUMBER : ')
    model = input()
    print('ENTER THE QUANTITY : ')
    quantity = int(input())
    print('ENTER THE PLACE OF STORAGE : ')
    place = input()
    product_check=sales_inventory_check(brand,model,place,quantity)
    if len(product_check)==0:
        print('ITEM IS NOT AVAILABLE IT CANNOT BE ADDED TO SALES')
    else : 
        print('ENTER THE SELLING PRICE : ')
        selling_price=int(input())
        bdate = get_date()
        btime = get_time()
        product_code = int(product_check['PRODUCT CODE'])
        billing_code=billing_code_generator(name,quantity,product_code)
        entry = [product_code , name , brand , model , quantity , place , bdate , btime , selling_price , billing_code]
        sales_csv.loc[len(sales_csv)] = entry
        save_sales_of_csv(sales_csv)
        quantity_changer('subtract',quantity,product_code)
def return_product():
    sales_csv=open_sales_of_csv()
    print('ENTER THE COSTUMER NAME : ')
    name=input()
    print('ENTER THE BILLING CODE : ')
    billing_code=input().upper()
    product_code=int(sales_csv[sales_csv['BILLING CODE']==billing_code]['PRODUCT CODE'])
    quantity=int(sales_csv[sales_csv['BILLING CODE']==billing_code]['QUANTITY'])
    quantity_changer('add',quantity,product_code)
    sales_csv=sales_csv[sales_csv['BILLING CODE']!=billing_code]
    save_sales_of_csv(sales_csv)
```

```python
def sales_manager():
    choice=1
    while(choice<3):
        choice=sales_manager_menu()
        if choice>3:
            print('ENTER A VALID CHOICE')
        if choice==1:
            add_sales()
        elif choice==2:
            return_product()
        elif choice==3:
            break
```

```python
#inventory_veiw function
def inventory_view():
    inventory_csv=open_inventory_of_csv()
    inventory_display=inventory_csv[['BRAND','MODEL','PLACE OF STORAGE','QUANTITY']]
    print(inventory_display)
```

```python
#sales_stat inside functions
def sales_stat_menu():
    print('1-TODAY')
    print('2-LAST DAY')
    print('3-LAST WEEK')
    print('4-LAST MONTH')
    print('5-CUSTOMIZED VIEW')
    print('6-MAIN MENU')
    print('ENTER YOUR CHOICE : ')
    choice = int(input())
    return choice
def tday():
    sales_csv=open_sales_of_csv()
    today = date.today()
    d = today.strftime("%b-%d-%Y")
    data=sales_csv[sales_csv['DATE']==d]
    print(data)
    
def last_day():
    sales_csv=open_sales_of_csv()
    today = date.today()
    yesterday = today - timedelta(days = 1)
    d = yesterday.strftime("%b-%d-%Y")
    data=sales_csv[sales_csv['DATE']==d]
    print(data)
    
def last_week():
    sales_csv=open_sales_of_csv()
    today = date.today()
    s = today - timedelta(days = 7)
    e = today - timedelta(days = 1)
    start = s.strftime("%b-%d-%Y")
    end= e.strftime("%b-%d-%Y")
    data=sales_csv[sales_csv['DATE']>=start]
    data=data[data['DATE']<=end]
    print(data)
    
def last_month():
    sales_csv=open_sales_of_csv()
    today = date.today()
    s = today - timedelta(days = 30)
    e = today - timedelta(days = 1)
    start = s.strftime("%b-%d-%Y")
    end= e.strftime("%b-%d-%Y")
    data=sales_csv[sales_csv['DATE']>=start]
    data=data[data['DATE']<=end]
    print(data)
     
def customize():
    sales_csv=open_sales_of_csv()
    print('ENTER THE NUMBER OF DAYS : ')
    num=int(input())
    today = date.today()
    s = today - timedelta(days = num)
    e = today - timedelta(days = 1)
    start = s.strftime("%b-%d-%Y")
    end= e.strftime("%b-%d-%Y")
    data=sales_csv[sales_csv['DATE']>=start]
    data=data[data['DATE']<=end]
    print(data)
    
    
```

```python
# sales_stats function
def sales_stat():
    choice=1
    while(choice<7):
        choice=sales_stat_menu()
        if choice>6:
            print('ENTER A VALID CHOICE')
        if choice==1:
            tday()
        elif choice==2:
            last_day()
        elif choice==3:
            last_week()
        elif choice==4:
            last_month()
        elif choice==5:
            customize()
        elif choice==6:
            break
```

```python
def alert_menu():
    print('1-ALERTS')
    print('2-CUSTOMIZED ALERTS')
    print('3-MAIN MENU')
    print('ENTER YOUR CHOICE : ')
    choice = int(input())
    return choice
def default_alert():
    inventory_csv=open_inventory_of_csv()
    alert=2
    data=inventory_csv[inventory_csv['QUANTITY']<=alert]
    print('STOCKS WITH LOW QUANTITY - ')
    print(data)
def customized_alert():
    inventory_csv=open_inventory_of_csv()
    print('ENTER THE LIMIT : ')
    alert=int(input())
    data=inventory_csv[inventory_csv['QUANTITY']<=alert]
    print('STOCKS WITH LOW QUANTITY - ')
    print(data)
```

```python
def alerts():
    choice=1
    while(choice<4):
        choice=alert_menu()
        if choice>3:
            print('ENTER A VALID CHOICE')
        if choice==1:
            default_alert()
        elif choice==2:
            customized_alert()
        elif choice==3:
            break
```
